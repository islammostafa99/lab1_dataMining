we read the data from the file in google drive and parse its attributes by ',' and store it in array using pandas library.

```python
import pandas as pd
file_path = 'drive/MyDrive/Untitled folder/wdbc.data'
array=pd.read_csv(file_path,sep=',',header=None)
array.head()
```

```python
cols = ["id","diagnosis","mean1","mean2","mean3","mean4","mean5","mean6","mean7","mean8","mean9","mean10","SE1","SE2","SE3","SE4","SE5","SE6","SE7","SE8","SE9","SE10","worst1","worst2","worst3","worst4","worst5","worst6","worst7","worst8","worst9","worst10"]
array.columns=cols
array.head()
```

Count plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

#count plot
sns.countplot(x = 'diagnosis', data = array)
plt.xlabel("diagnosis")
plt.ylabel("count")
plt.title("Count Plot");
plt.show()
```

Histogram for means

```python
M = array[array.diagnosis=='M']
B = array[array.diagnosis=='B']
M = M.iloc[:, 2:12].copy()
B = B.iloc[:, 2:12].copy()

for i in range (0,10) :
    M.iloc[:,i].hist(figsize=(12,4),bins=30,facecolor='r', label='M',alpha=0.7)
    B.iloc[:,i].hist(figsize=(12,4),bins=30,facecolor='b', label='B',alpha=0.7)
    plt.legend(loc='upper right')
    plt.title('Mean Histograms feature number {}'.format(i))
    plt.show();
```

Histogram for SE

```python
M= array[array.diagnosis=='M']
B= array[array.diagnosis=='B']
M = M.iloc[:, 12:22].copy()
B = B.iloc[:, 12:22].copy()

for i in range (0,10) :
    M.iloc[:,i].hist(figsize=(12,4),bins=30,facecolor='r', label='M',alpha=0.7)
    B.iloc[:,i].hist(figsize=(12,4),bins=30,facecolor='b', label='B',alpha=0.7)
    plt.legend(loc='upper right')
    plt.title('SE Histograms feature number {}'.format(i-1))
    plt.show();
```

Histogram for worst

```python
M= array[array.diagnosis=='M']
B= array[array.diagnosis=='B']
M = M.iloc[:, 22:32].copy()
B = B.iloc[:, 22:32].copy()

for i in range (0,10) :
    M.iloc[:,i].hist(figsize=(12,4),bins=30,facecolor='r', label='M',alpha=0.7)
    B.iloc[:,i].hist(figsize=(12,4),bins=30,facecolor='b', label='B',alpha=0.7)
    plt.legend(loc='upper right')
    plt.title('worst Histograms feature number {}'.format(i-1))
    plt.show();
```

Boxplot for B

```python
B = array[array.diagnosis=='B'].iloc[:,2:32].copy()
B.head()
```

```python
for column in B.columns:
    B[column] = (B[column] - B[column].min()) / (B[column].max() - B[column].min())

B.head()
```

```python
B.boxplot(return_type='axes',figsize=(20,10))
```
